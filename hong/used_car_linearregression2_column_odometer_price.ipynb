# 4. 컬럼별 주행거리와 가격의 평균값 확인
- type, manufacturer, cylinder, fuel, transmission, drive, paint_color
- 컬럼과 가격과의 관계에서 주행거리가 어떤 변화가 있는지, 직관적으로 확인

### 4-1. type
- truck, pickup, offroad가 평균적으로 비싸다.
- 주행거리가 크다고해서 가격이 싼것은 아니다.
- hatchback은 주행거리가 작은편인데 가격이 제일 싼 편이다.
- convertible, coupe 는 주행거리가 작고 가격이 비싼 편이다.

```python
df_type_mean = df.groupby("type").agg("mean").reset_index()[["type", "odometer", "price"]]
df_type_mean.sort_values(["price"], ascending=True, inplace=True)
df_type_mean
```

### 4-2. manufacturer
- ferrari 가 평균가격이 제일 비싸다.
- estonmartin, morgan 도 평균가격이 비싸다.
- tesla 는 상위 3개 브랜드 보다 주행거리가 2배 높은데도 평균 가격이 높은편이다.
- ford, gmc, ram 은 주행거리가 10만이 넘어도, 평균가격은 높은 편이다.

```python
df_manufacturer_mean = df.groupby("manufacturer").agg("mean").reset_index()[["manufacturer", "odometer", "price"]]
df_manufacturer_mean.sort_values(["price"], ascending=True, inplace=True)

pd.concat([df_manufacturer_mean, df_manufacturer], axis=1).sort_values(["price"], ascending=True)
```

```python
df.loc[df["manufacturer"] == "morgan", :]
```

```python
df.loc[df["manufacturer"] == "tesla", :]
```

### 4-3. cylinders
- 12 cy가 평균가격이 가장 비싸다.
- oters 가 주행거리고 작은편이고, 평균가격이 비싸다.
- 8, 10 cy는 주행거리가 큰 편인데 평균가격은 비싸다.

```python
df_cylinders_mean = df.groupby("cylinders").agg("mean").reset_index()[["cylinders", "odometer", "price"]]
df_cylinders_mean.sort_values(["price"], ascending=True, inplace=True)

pd.concat([df_cylinders_mean, df_cylinders], axis=1).sort_values(["price"], ascending=True)
```

```python
df.loc[df["cylinders"] == "other", :]
```

```python
df.loc[df["cylinders"] == "3 cylinders", :]
```

### 4-4. fuel
- disel 이 주행거리가 크지만 평균가격이 높은 편이다.
- electric 은 갯수가 적으면서 평균가격이 높은 편이다.
- hybrid 는 가격이 제일 낮다.

```python
df_fuel_mean = df.groupby("fuel").agg("mean").reset_index()[["fuel", "odometer", "price"]]
df_fuel_mean.sort_values(["price"], ascending=True, inplace=True)

pd.concat([df_fuel_mean, df_fuel], axis=1).sort_values(["price"], ascending=True)
```

```python
df.loc[df["fuel"] == "electric", :]
```

```python
df.loc[df["fuel"] == "hybrid", :]
```

### 4-5. transmission
- other 가 평균 가격이 높다.
- automatic 이 거의 모든 차종에 해당한다. 
- manual 은 갯수가 적고 평균 가격도 낮은편이다.

```python
df_transmission_mean = df.groupby("transmission").agg("mean").reset_index()[["transmission", "odometer", "price"]]
df_transmission_mean.sort_values(["price"], ascending=True, inplace=True)

pd.concat([df_transmission_mean, df_transmission], axis=1).sort_values(["price"], ascending=True)
```

### 4-6. drive
- 4wd 가 갯수도 많고 평균 가격도 높다.
- rwd 는 갯수는 작지만, 가격은 높은 편이다. 
- fwd 는 평균 가격이 낮은 편이다.

```python
df_drive_mean = df.groupby("drive").agg("mean").reset_index()[["drive", "odometer", "price"]]
df_drive_mean.sort_values(["price"], ascending=True, inplace=True)

pd.concat([df_drive_mean, df_drive], axis=1).sort_values(["price"], ascending=True)
```

### 4-7. paint_color
- white 가 갯수도 많고, 주행거리도 크고 평균 가격이 높다.
- black 도 갯수도 많고, 주행거리도 크고 평균 가격이 높다.
- orange 는 갯수가 적고, 평균 가격이 높다. 
- green 은 갯수도 적고, 평균가격이 낮다.

```python
df_paint_color_mean = df.groupby("paint_color").agg("mean").reset_index()[["paint_color", "odometer", "price"]]
df_paint_color_mean.sort_values(["price"], ascending=True, inplace=True)

pd.concat([df_paint_color_mean, df_paint_color], axis=1).sort_values(["price"], ascending=True)
```
